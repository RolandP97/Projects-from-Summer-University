```python
df2_numeric = df2.select_dtypes(include='float64') #new variable with only columns that have numbers
df2_numeric.head()
```

```python
df2_numeric_column_names = df2_numeric.columns[:-1] #what does the -1 do?
df2_numeric_column_names
df2_numeric.head()
```

```python
df2_numeric_ndarray = df2_numeric[df2_numeric_column_names].to_numpy()
df2_numeric_ndarray #how to get normal looking output, instead of the scientific one?
```

```python
from sklearn import preprocessing
```

```python
#MINMAX

min_max_scaler = preprocessing.MinMaxScaler() 

df2_numeric_ndarray_minmax = min_max_scaler.fit_transform(df2_numeric_ndarray)
#scales the data to have mean of about 0 and variance unit of 1
df2_numeric_ndarray_minmax
```

```python
min_max_scaler.inverse_transform(df2_numeric_ndarray_minmax) #just puts the data back to original form
```

```python
g = sns.PairGrid(
    pd.DataFrame(df2_numeric_ndarray_minmax, columns=df2_numeric_column_names)
)
g.map_upper(plt.scatter, color="RoyalBlue", s=8)
g.map_lower(sns.kdeplot, color="Red")
g.map_diag(sns.distplot, color="Green")
```

```python
df2_numeric_minmax = pd.DataFrame(
    df2_numeric_ndarray_minmax, columns=df2_numeric_column_names
)
df2_numeric_minmax.head()  #creates a dataframe from the recently transformed data, giving it column names aswell
```

```python
#Z-score processing

zscore_scaler = preprocessing.StandardScaler()

df2_numeric_ndarray_zscore = zscore_scaler.fit_transform(df2_numeric_ndarray)

df2_numeric_ndarray_zscore
```

```python
g = sns.PairGrid(
    pd.DataFrame(df2_numeric_ndarray_zscore, columns=df2_numeric_column_names)
)
g.map_upper(plt.scatter, color="RoyalBlue", s=8)
g.map_lower(sns.kdeplot, color="Red")
g.map_diag(sns.distplot, color="Green")
```
